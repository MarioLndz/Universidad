```python
def DiferenciasDivididas (x, y):
    n = len(x)
    A = np.zeros([n, n])
    
    for i in range (0, n):
        A[i, 0] = y[i]
    
    for j in range (1, n):
        for i in range (j, n):
            A[i, j] = (A[i,j-1] - A[i-1,j-1])/(x[i] - x[i-j])
            
    return (A)
```

```python
def puntos(n = 14):
    x = np.zeros([n+1])

    for k in range (0, n+1):
        x[k] = -5 + ((10*k)/n)
    return (x)
```
